```python
from __future__ import print_function, division
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import math
import sys 
import time
sys.path.insert(0,'..')
from utils import plot_stroke
%matplotlib inline
```

```python
state_size = 300
num_classes = 121
num_layers = 3
features = 3
dropout_pkeep = 0.8
gradient_clip = 10
total_mixtures = 19
```

```python
batch_size = 3
seqlength = 50
```

```python
def normal1d(x,mu,sig):
    x_mu = tf.subtract(x, mu)
    return tf.square(tf.divide(x_mu, sig))

def Z(x1, x2, mu1, mu2, sig1, sig2, rho):
        x11 = normal1d(x1,mu1,sig1)
        x12 = normal1d(x2,mu2,sig2)
        x1x2 = tf.divide((tf.multiply(np.subtract(x1, mu1),tf.subtract(x2, mu2))),(tf.multiply(sig1,sig2)))
        Z = x11+x12-(2*tf.multiply(rho,x1x2))
        return Z

def gaussian2d_my(x1, x2, mu1, mu2, s1, s2, rho):
    z = Z(x1, x2, mu1, mu2, s1, s2, rho)
    rho_sqr = 1-tf.square(rho)
    expo = tf.exp(tf.div(-z,(2*(rho_sqr))))
    denominator = tf.reciprocal(2*np.pi*np.multiply(np.multiply(s1, s2), tf.sqrt(rho_sqr)))
    result = tf.multiply(expo,denominator)
    return result

def get_loss_my(pi, coordinates, mu1, mu2, sigma1, sigma2, rho, eos):
    [x3,x1, x2] = coordinates
    gaussian = gaussian2d_my(x1, x2, mu1, mu2, sigma1, sigma2, rho)
    normal_loss1 = tf.multiply(gaussian, pi)
    normal_loss2 = tf.reduce_sum(normal_loss1,1,keep_dims=True) 
    normal_loss2 = tf.maximum(normal_loss2, 1e-20)
    normal_loss3 = -tf.log(normal_loss2) 
    i = tf.constant(0)
    values_eos =  tf.Variable([],dtype=tf.float32)

    def condition(i,values_eos):
        return i<(batch_size*seqlength)
    
    def body(i,values_eos):
        result = tf.cond(x3[i][0] < 1, lambda:-tf.log(1-eos[i]), lambda: -tf.log(eos[i]))
        values_eos = tf.concat([values_eos,result],0)
        return tf.add(i, 1),values_eos

    # do the loop:
    r,eos_result_log  = tf.while_loop(condition, body, [i,values_eos],shape_invariants=[i.get_shape(),
                                                   tf.TensorShape([None])])


    eos_result_log = tf.reshape(eos_result_log,tf.shape(normal_loss3))
    result = tf.reduce_sum(normal_loss3+eos_result_log)
    
    return result,normal_loss3,eos_result_log,gaussian
                      
def get_mdn_coef_my(mixture):


        eos = mixture[:, 0:1]
        mixture = mixture[:, 1:]

        #Same order as paper
        pi, mu1, mu2, sigma1, sigma2, rho = tf.split(mixture, 6, 1)              
        #equation 18
        eos = tf.reciprocal(1+tf.exp(eos))              
        #equation 19
        pi = tf.nn.softmax(pi)
        #equation 20
        #No Change
        #equation 21
        sigma1 = tf.exp(sigma1)
        sigma2 = tf.exp(sigma2)
        #equation 22
        rho = tf.tanh(rho)

        return [eos, pi, mu1, mu2, sigma1, sigma2, rho]
                      
```

```python
weight_noise = tf.truncated_normal_initializer(stddev=0.075, mean=0.0)
X = tf.placeholder(tf.float32, [None, None,features]) # [ batch_size, SEQLEN,features ]
Y_pred = tf.placeholder(tf.float32, [None,features ]) # [ batch_size, SEQLEN,features ]
init_state = tf.placeholder(tf.float32, [num_layers, 2, None, state_size])


W2 = tf.Variable(np.random.rand(state_size, total_mixtures),dtype=tf.float32)
b2 = tf.Variable(np.zeros((1,total_mixtures)), dtype=tf.float32)

```

```python
state_per_layer_list = tf.unstack(init_state, axis=0)
rnn_tuple_state = tuple(
            [tf.contrib.rnn.LSTMStateTuple(state_per_layer_list[idx][0], state_per_layer_list[idx][1])
             for idx in range(num_layers)]
        )


stacked_rnn = []
for _ in range(num_layers):
    cell = tf.nn.rnn_cell.LSTMCell(state_size)
    cell = tf.nn.rnn_cell.DropoutWrapper(cell, output_keep_prob=dropout_pkeep)
    stacked_rnn.append(cell)

MultiRNNCell = tf.nn.rnn_cell.MultiRNNCell(stacked_rnn , state_is_tuple=True)
states_series,current_state = tf.nn.dynamic_rnn(cell= MultiRNNCell,inputs= X, initial_state= rnn_tuple_state)
# states_series: [ batch_size, SEQLEN, state_size ]
# current_state  [ batch_size, state_size*num_layers ]



```

```python
strokes = np.load('../data/strokes.npy')
texts = ""
with open('../data/sentences.txt') as f:
    texts = f.readlines()
```

```python
states_series = tf.reshape(states_series, [-1, state_size])
logits = tf.matmul(states_series, W2) + b2 
coordinates = tf.split(Y_pred, features, 1) 


[eos, pi, mu1, mu2, sigma1, sigma2, rho]  = get_mdn_coef_my(logits)
loss,_,_,_ = get_loss_my(pi, coordinates, mu1, mu2, sigma1, sigma2, rho, eos)

# Normalize loss over batch
loss = tf.divide(loss,batch_size)

# https://stackoverflow.com/questions/36498127/how-to-effectively-apply-gradient-clipping-in-tensor-flow

optimizer = tf.train.AdamOptimizer(0.002)
gradients, variables = zip(*optimizer.compute_gradients(loss))
gradients, _ = tf.clip_by_global_norm(gradients, gradient_clip)
optimize = optimizer.apply_gradients(zip(gradients, variables))

# Add ops to save and restore all the variables.
saver = tf.train.Saver()
```

```python
def sample_gaussian2d(_mu1, _mu2, _s1, _s2, _rho):
    mean = [_mu1, _mu2]
    cov = [[_s1*_s1, _rho*_s1*_s2], [_rho*_s1*_s2, _s2*_s2]]
    _x = np.random.multivariate_normal(mean, cov, 1)
    return _x[0][0], _x[0][1]

```

```python
def get_pi_idx(x, pdf):
  N = pdf.size
  accumulate = 0
  for i in range(0, N):
    accumulate += pdf[i]
    if (accumulate >= x):
      return i
  print 'error with sampling ensemble'
  return -1
```
